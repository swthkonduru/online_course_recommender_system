```python
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
data = {'KNN':0.20747353396942664,
        'NMF':0.20585305681684202,
        'Elasticnet':0.21200026329107982,
        'Ridge':0.21132863543056202,
        'Lasso':0.21200026329107982,
        'StackingClassifier':0.09820296087700026,
        'Neutral_Network':0.23890420794487}
```

```python

x = list(data.keys())
y = list(data.values())

plt.figure(figsize=(20,10))
sb.set(font_scale=1.1, style="whitegrid")
sb.barplot(x=x, y=y, palette='magma')

plt.ylabel('RMSE', fontsize=15, fontweight='bold')
plt.title('Model RMSE Comparison', fontsize=15, fontweight='bold')
plt.show()
```
